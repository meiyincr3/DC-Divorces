# ANÁLISIS
La distribución combinada de las edades en el momento del divorcio muestra que la mayoría de los divorcios ocurren cuando los individuos tienen entre 30 y 45 años.

# ANÁLISIS
La distribución muestra que las edades más comunes en el momento del matrimonio están entre los 20 y 35 años.

```python
merged_data = pd.merge(dEdadMat, dEdadDivorce, on='edad', how='outer').fillna(0)
```

```python
merged_data = merged_data.rename(columns={'total_x': 'total_mat', 'total_y': 'total_div'})
merged_data.to_csv("../PROYECTO P1/datos.csv", index=False)
merged_data
```

```python
correlacion = merged_data['total_mat'].corr(merged_data['total_div'])
correlacion
```

```python
# Definir categorías de edad
age_bins = [10, 20, 30, 40, 50, 60, 70, 80, 90]
merged_data['age_group'] = pd.cut(merged_data['edad'], bins=age_bins)
merged_data
```

```python
# Calcular la tasa de divorcio por grupo de edad en el momento del matrimonio
divorce_rate_by_marriage_age = merged_data.groupby('age_group').agg({'total_mat': 'sum', 'total_div': 'sum'})
divorce_rate_by_marriage_age['divorce_rate'] = divorce_rate_by_marriage_age['total_div'] / divorce_rate_by_marriage_age['total_mat']
divorce_rate_by_marriage_age
```

```python
# Gráfico de la tasa de divorcio por grupo de edad en el momento del matrimonio
plt.figure(figsize=(10, 6))
divorce_rate_by_marriage_age['divorce_rate'].plot(kind='bar')
plt.title('Tasa de divorcio por grupo de edad en el momento del matrimonio')
plt.xlabel('Grupo de edad')
plt.ylabel('Tasa de divorcio')
plt.xticks(rotation=45)
plt.show()
```
