```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv("clean_divorce.csv", sep = ",")
df.head()
```

## ¿Cuál es la relación entre la duración del matrimonio y el número de hijos que viven dentro del hogar? ¿Este factor provoca el divorcio? 

```python
# Supuesto: Asumimos que el número total de hijos es el máximo entre hijos_1 e hijos_2,
# ya que es posible que algunos hijos sean compartidos entre ambos cónyuges.
df['total_hijos'] = df[['hijos_1', 'hijos_2']].max(axis=1)
```

```python
# Scatter plot para visualizar la relación entre duración del matrimonio y el número de hijos
plt.figure(figsize=(10, 6))
sns.scatterplot(x='dur_mat', y='total_hijos', data=df, hue = 'total_hijos', palette = "Set1")
plt.title('Relación entre la duración del matrimonio y el número de hijos')
plt.xlabel('Duración del matrimonio (años)')
plt.ylabel('Número total de hijos')
plt.legend(title='Total Hijos')
plt.savefig('relacion_matrimonio_hijos.png')
plt.show()
```

El gráfico de dispersión actual proporciona una primera impresión de la relación entre la duración del matrimonio y el número de hijos, pero tiene limitaciones en la interpretación debido a la superposición de datos, la falta de tendencias claras y la alta densidad de puntos en ciertas áreas. Para un análisis más profundo y preciso, se recomienda explorar otro tipo de gráficos adicionales como boxplots y realizar pruebas estadísticas que permitan evaluar la relación entre estas variables de manera más completa y detallada.

```python
# Boxplots para analizar las distribuciones de las duraciones de los matrimonios por el número de hijos en el hogar.
plt.figure(figsize=(10, 7))
boxplot = sns.boxplot(hue='total_hijos', y='dur_mat', data=df, palette='Set1')
plt.title('Distribución de la duración del matrimonio por número de hijos')
plt.xlabel('Número Total de hijos')
plt.ylabel('Duración del matrimonio (años)')
plt.legend(title='Total Hijos')
plt.savefig('distribucion_boxplot_matrimonio_hijos.png')
plt.show()
```

El gráfico de boxplot revela varias tendencias interesantes en la relación entre la duración del matrimonio y el número de hijos. Las parejas sin hijos tienden a tener matrimonios más largos, mientras que aquellas con 1 a 4 hijos muestran una menor duración del matrimonio. Curiosamente, las parejas con 5 hijos tienen una mayor mediana de duración del matrimonio, lo que sugiere que podrían existir factores adicionales en juego. Es importante notar que solo existe una pareja con 7 hijos, lo que limita la interpretación para este grupo específico. Este análisis sugiere que la relación entre el número de hijos y la duración del matrimonio puede no ser lineal y podría estar influenciada por una variedad de factores contextuales y personales.

```python
correlation, p_value = stats.pearsonr(df['dur_mat'], df['total_hijos'])
print(f'Correlación entre duración del matrimonio y número de hijos: {correlation:.2f}')
print(f'Valor p: {p_value:.4f}')

# Contraste de hipótesis
# Ho: No existe una relación lineal entre la duración del matrimonio y el número de hijos. (p=0)
# Ha: Existe una relación lineal entre la duración del matrimonio y el número de hijos. (p != 0)
if p_value < 0.05:
    print("La correlación es estadísticamente significativa.")
else:
    print("La correlación no es estadísticamente significativa.")
```

El análisis de correlación de Pearson realizado sobre los datos de duración del matrimonio y el número de hijos ha arrojado una correlación de -0.21 con un valor p significativo (p < 0.05). Este resultado indica que existe una relación lineal débil negativa entre la duración del matrimonio y el número de hijos, lo que significa que, estadísticamente, a medida que aumenta el número de hijos, la duración del matrimonio tiende a disminuir ligeramente. No obstante, debido a la debilidad de esta correlación, es probable que otros factores influyan más significativamente en la duración del matrimonio. Además, dado que el coeficiente de correlación es relativamente bajo pero el valor p es significativo, esto sugiere que podría existir una relación no lineal entre estas variables que no ha sido capturada adecuadamente por el análisis de correlación lineal de Pearson.

## ¿Cuál es la probabilidad de que un matrimonio en Ecuador sobreviva más de 30 años, basándonos en el análisis de supervivencia?

Esta pregunta busca determinar la probabilidad de que los matrimonios en Ecuador duren más de 30 años antes de disolverse, utilizando las técnicas de análisis de supervivencia descritas en el paper:  Raed Abdel-Sater (2022): Marriage Dissolution in the United States: A Survival
 Analysis Approach, Journal of Divorce & Remarriage, DOI: 10.1080/10502556.2022.2042788 (link: https://doi.org/10.1080/10502556.2022.2042788).

```python
from lifelines import KaplanMeierFitter
```

```python
# Creacion del dataframe survival para el análisis de probabilidad
df_survival = df[['dur_mat']].copy()
df_survival['event'] = 1  # Todos los registros son divorcios

# Histograma de la duración del matrimonio
plt.figure(figsize=(10, 7))
plt.hist(df_survival['dur_mat'], bins=range(0, 75, 5), color='blue', edgecolor='red')
plt.title('Distribución de la Duración del Matrimonio en Ecuador')
plt.xlabel('Duración del Matrimonio (años)')
plt.ylabel('Cantidad de Divorcios')
plt.axvline(x=30, color='r', linestyle='--', label='30 años')
plt.legend()
plt.show()
```

El gráfico proporciona una visión clara de cómo la duración del matrimonio está relacionada con la probabilidad de divorcio en Ecuador. La alta frecuencia de divorcios en los primeros años de matrimonio destaca la importancia de los primeros años para la estabilidad matrimonial. La disminución en la frecuencia de divorcios después de los 30 años sugiere que los matrimonios más largos tienden a ser más estables y duraderos. Este análisis puede ser útil para estudios sociológicos y para diseñar intervenciones destinadas a mejorar la estabilidad matrimonial en los primeros años de casamiento.

```python
# Gráfico de líneas para mostrar la cantidad de divorcios a lo largo de los años
divorce_counts = df_survival['dur_mat'].value_counts().sort_index()
plt.figure(figsize=(10, 7))
plt.plot(divorce_counts.index, divorce_counts.values, marker='o', color='darkred')
plt.title('Cantidad de Divorcios a lo Largo de los Años')
plt.xlabel('Duración del Matrimonio (años)')
plt.ylabel('Cantidad de Divorcios')
plt.axvline(x=30, color='r', linestyle='--', label='30 años')
plt.legend()
plt.show()
```

El gráfico de líneas muestra la cantidad de divorcios a lo largo de los años de matrimonio. Al igual que en el histograma anterior, se observa que la mayoría de los divorcios ocurren en los primeros 30 años de matrimonio, con un pico notable entre los 5 y 10 años. Después de los 30 años, la cantidad de divorcios disminuye significativamente, indicando una mayor estabilidad en matrimonios que superan esta duración. Este análisis refuerza la observación de que los primeros años de matrimonio son críticos para la estabilidad matrimonial, mientras que los matrimonios que duran más de 30 años tienen una probabilidad significativamente menor de disolverse.

```python
# Gráfico de densidad de los divorcios
plt.figure(figsize=(10, 7))
sns.kdeplot(df_survival['dur_mat'], fill=True, color='purple')
plt.title('Densidad de los Divorcios según la Duración del Matrimonio')
plt.xlabel('Duración del Matrimonio (años)')
plt.axvline(x=30, color='r', linestyle='--', label='30 años')
plt.legend()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(25, 9))

# Histograma de la duración del matrimonio
# axs[0].hist(df_survival['dur_mat'], bins=range(0, 75, 5), color='blue', edgecolor='red')
# axs[0].set_title('Distribución de la Duración del Matrimonio en Ecuador')
# axs[0].set_xlabel('Duración del Matrimonio (años)')
# axs[0].set_ylabel('Cantidad de Divorcios')
# axs[0].axvline(x=30, color='r', linestyle='--', label='30 años')
# axs[0].legend()

# Gráfico de cantidad de divorcios a lo largo de los años
divorce_counts = df_survival['dur_mat'].value_counts().sort_index()
axs[0].plot(divorce_counts.index, divorce_counts.values, marker='o', color='darkred')
axs[0].set_title('Cantidad de Divorcios a lo Largo de los Años')
axs[0].set_xlabel('Duración del Matrimonio (años)')
axs[0].set_ylabel('Cantidad de Divorcios')
axs[0].axvline(x=30, color='r', linestyle='--', label='30 años')
axs[0].legend()

# Gráfico de densidad
sns.kdeplot(df_survival['dur_mat'], fill=True, color='purple', ax=axs[1])
axs[1].set_title('Densidad de los Divorcios según la Duración del Matrimonio')
axs[1].set_xlabel('Duración del Matrimonio (años)')
axs[1].set_ylabel('Densidad de Divorcios')
axs[1].axvline(x=30, color='r', linestyle='--', label='30 años')
axs[1].legend()

plt.tight_layout()
plt.savefig('divorcios_subplots.png')
plt.show()
```

```python
# KaplanMeierFitter
# El estimador de Kaplan-Meier, también conocido como estimador de producto-límite,
# se utiliza para calcular la probabilidad de supervivencia en distintos puntos de tiempo.
# Modelo paramétrico usado en el paper:
kmf = KaplanMeierFitter()

# Ajustar el modelo de Kaplan-Meier y graficar la función de supervivencia
plt.figure(figsize=(12, 8))

# Ajuste del modelo
kmf.fit(durations=df_survival['dur_mat'], event_observed=df_survival['event'], label='Supervivencia del Matrimonio')

# F+inción de supervivencia del matrimonio en Ecuador
kmf.plot_survival_function()
plt.title('Función de Supervivencia del Matrimonio en Ecuador')
plt.xlabel('Duración del Matrimonio (años)')
plt.ylabel('Probabilidad de Supervivencia')
plt.axvline(x=30, color='r', linestyle='--', label='30 años')
plt.legend()
plt.savefig('funcion_supervivencia.png')
plt.show()
```

```python
# Resultado de la probabilidad de supervivencia a 30 años
prob_30_years = kmf.survival_function_at_times(30).values[0]
print(f'Probabilidad de que un matrimonio dure más de 30 años en Ecuador: {prob_30_years:.4f}')
```

El modelo de Kaplan-Meier revela que la supervivencia de los matrimonios en Ecuador decrece significativamente con el tiempo. La alta tasa de divorcios en los primeros años de matrimonio sugiere que estos años son críticos para la estabilidad matrimonial. A partir de los 30 años, la probabilidad de supervivencia disminuye notablemente, con solo el 13.12% de los matrimonios alcanzando o superando esta duración. Este análisis estadístico proporciona una visión sobre los patrones de disolución matrimonial en Ecuador y subraya la importancia de los primeros años en la vida matrimonial.

Este tipo de análisis puede ser útil para diseñar políticas y programas de intervención dirigidos a apoyar a las parejas en los primeros años de matrimonio, con el objetivo de aumentar la estabilidad matrimonial y reducir las tasas de divorcio.
