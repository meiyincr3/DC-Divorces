```python
import seaborn as sb
import pandas as pd
import geopandas as gpd

import plotly.express as px
```

```python
df = pd.read_csv("../input/inec_matrimonios_2023.csv", sep=';')
```

```python
df
```

```python
df.fecha_insc = pd.to_datetime(df.fecha_insc)
```

```python
df.columns
```

```python
df = df.loc[((df.cod_pais1 == "Ecuador") & (df.cod_pais2 == "Ecuador")),:]
```

```python
df = df.replace('Sin información', pd.NA)
```

```python
df.dropna(inplace=True)
```

```python
trashColumns = ["anio_insc","mes_insc","dia_insc","anio_insc","anio_nac1","mes_nac1"
               ,"dia_nac1","fecha_nac1","fecha_nac2","anio_nac2","mes_nac2","dia_nac2","niv_inst1"
               ,"niv_inst2",'nmatant1', 'est_civi1', 'p_etnica1', 'sabe_leer1', 'parr_hab1', 
                'nmatant2', 'est_civi2', 'p_etnica2', 'parr_hab2','hijos_rec', 'sabe_leer2', "cod_pais1", "cod_pais2"]
df.drop(trashColumns, axis=1, inplace=True)
```

```python
df
```

```python
df.to_csv("../output/clean_matrimonios.csv",index=False)
```

```python
persona1xprov = pd.pivot_table(df, index='prov_hab1', columns='sexo_1', aggfunc='size', fill_value=0).reset_index()
```

```python
persona2xprov = pd.pivot_table(df, index='prov_hab2', columns='sexo_2', aggfunc='size', fill_value=0).reset_index()
```

```python
cantxprov = persona1xprov.merge(persona2xprov, left_on="prov_hab1", right_on ="prov_hab2")
```

```python
cantxprov.columns = cantxprov.columns.str.strip()
```

```python
cantxprov["hombresCasados"] = cantxprov["Hombre_x"]+cantxprov["Hombre_y"]
```

```python
cantxprov["mujeresCasadas"] = cantxprov["Mujer_x"]+cantxprov["Mujer_y"]
```

```python
cantxprov = cantxprov[["prov_hab1", "hombresCasados", 'mujeresCasadas']]
```

```python
cantxprov["totalMatrimonios"] = cantxprov.hombresCasados + cantxprov.mujeresCasadas
```

```python
cantxprov.loc[:, 'prov_hab1']=cantxprov["prov_hab1"].str.lower()
```

```python
cantxprov.loc[:, 'prov_hab1'] = cantxprov.prov_hab1.str.replace("santo domingo de los tsáchilas","santo domingo de los tsachilas")
cantxprov.loc[:, 'prov_hab1'] = cantxprov.prov_hab1.str.replace("manabí","manabi")
cantxprov.loc[:, 'prov_hab1']= cantxprov.prov_hab1.str.replace("los ríos","los rios")
cantxprov.loc[:, 'prov_hab1']= cantxprov.prov_hab1.str.replace("sucumbíos","sucumbios")
cantxprov.loc[:, 'prov_hab1'] = cantxprov.prov_hab1.str.replace("bolívar", "bolivar")
cantxprov.loc[:, 'prov_hab1'] = cantxprov.prov_hab1.str.replace("galápagos","galapagos")
cantxprov
```

```python
cantxprov.to_csv("../output/matrimonios2023.csv",index=False)
```
