# PREGUNTA 1
¿Existe una correlación significativa entre la tasa de divorcios y la tasa de matrimonios en las diferentes provincias de Ecuador?

```python
persona1xprov = pd.pivot_table(df, index='prov_hab1', columns='sexo_1', aggfunc='size', fill_value=0).reset_index()
```

```python
persona2xprov = pd.pivot_table(df, index='prov_hab2', columns='sexo_2', aggfunc='size', fill_value=0).reset_index()
```

```python
cantxprov = persona1xprov.merge(persona2xprov, left_on="prov_hab1", right_on ="prov_hab2")
```

```python
cantxprov.columns = cantxprov.columns.str.strip()
```

```python
cantxprov["hombresDivorciados"] = cantxprov["Hombre_x"]+cantxprov["Hombre_y"]
```

```python
cantxprov["mujeresDivorciadas"] = cantxprov["Mujer_x"]+cantxprov["Mujer_y"]
```

```python
cantxprov
```

```python
cantxprov = cantxprov[["prov_hab1", "hombresDivorciados", 'mujeresDivorciadas']]
```

```python
cantxprov["totalDivorcios"] = cantxprov.hombresDivorciados + cantxprov.mujeresDivorciadas
```

```python
cantxprov.dtypes
```

```python
cantxprov.loc[:, 'prov_hab1']=cantxprov["prov_hab1"].str.lower()
```

```python
cantxprov.rename(columns={'prov_hab1': 'provincia'}, inplace=True)
```

```python
cantxprov.loc[:, 'provincia'] = cantxprov.provincia.str.replace("santo domingo de los tsáchilas","santo domingo de los tsachilas")
cantxprov.loc[:, 'provincia'] = cantxprov.provincia.str.replace("manabí","manabi")
cantxprov.loc[:, 'provincia']= cantxprov.provincia.str.replace("los ríos","los rios")
cantxprov.loc[:, 'provincia']= cantxprov.provincia.str.replace("sucumbíos","sucumbios")
cantxprov.loc[:, 'provincia'] = cantxprov.provincia.str.replace("bolívar", "bolivar")
cantxprov.loc[:, 'provincia'] = cantxprov.provincia.str.replace("galápagos","galapagos")
```

```python
pf.rename(columns={'Provincia': 'provincia','Cantón': 'canton',"Total":"total"}, inplace=True)
```

```python
pf.provincia = pf.provincia.str.lower()
pf.canton = pf.canton.str.lower()
```

```python
pf.loc[:, 'provincia'] = pf.provincia.str.replace("santo domingo de los tsáchilas","santo domingo de los tsachilas")
pf.loc[:, 'provincia'] = pf.provincia.str.replace("manabí","manabi")
pf.loc[:, 'provincia'] = pf.provincia.str.replace("los ríos","los rios")
pf.loc[:, 'provincia'] = pf.provincia.str.replace("sucumbíos","sucumbios")
pf.loc[:, 'provincia'] = pf.provincia.str.replace("bolívar", "bolivar")
pf.loc[:, 'provincia'] = pf.provincia.str.replace("galápagos","galapagos")
```

```python
cantxprov
```

```python
mf.rename(columns={'prov_hab1': 'provincia'}, inplace=True)
```

```python
pf
```

```python
poblacionProv = pf.groupby('provincia').sum().reset_index()
```

```python
poblacionProv = poblacionProv[["provincia","total"]]
```

```python
poblacionProv
```

```python
divorcios_por_provincia = cantxprov.merge(poblacionProv, on='provincia')
matrimonios_por_provincia = mf.merge(poblacionProv, on='provincia')
```

```python
divorcios_por_provincia['tasa_divorcios'] = (divorcios_por_provincia['totalDivorcios'] / divorcios_por_provincia['total']) * 1000
```

```python
matrimonios_por_provincia['tasa_matrimonios'] = (matrimonios_por_provincia['totalMatrimonios'] / matrimonios_por_provincia['total']) * 1000
```

```python
divorcios_por_provincia
```

```python
comparacion = divorcios_por_provincia[['provincia', 'tasa_divorcios']].merge(
    matrimonios_por_provincia[['provincia', 'tasa_matrimonios']], on='provincia')
```

```python
comparacion
```

```python
correlacion = comparacion['tasa_matrimonios'].corr(comparacion['tasa_divorcios'])
correlacion
```

```python
fig, ax = plt.subplots(figsize=(6,4))
sb.scatterplot(x='tasa_matrimonios', y='tasa_divorcios', data=comparacion)
ax.set_title('Comparación de Tasas de Matrimonios y Divorcios por Provincia')
ax.set_xlabel('Tasa de Matrimonios (por 1000 personas)')
ax.set_ylabel('Tasa de Divorcios (por 1000 personas)')
```

# PREGUNTA 2
¿Las provincias con más matrimonios también tienen una mayor cantidad de divorcios en Ecuador?

```python
gprov.rename(columns={'dpa_despro': 'provincia'}, inplace=True)
```

```python
gprov = gprov.loc[gprov.provincia != "zona no delimitada" ,:]
```

```python
gprov.provincia = gprov.provincia.str.lower()
```

```python
gdp = gprov[["provincia","geometry"]]
```

```python
gdp = gdp.loc[gdp.provincia != "zona no delimitada" ,:]
```

```python
gdp
```

```python

gdp.set_index('provincia', inplace=True)
```

```python
# cantxprov.reset_index(inplace=True)
# mf.reset_index(inplace=True)
```

```python
mf
```

```python
divf = cantxprov.loc[cantxprov.provincia != "zona no delimitada" ,:]
```

```python
matf = mf.loc[mf.provincia != "exterior" ,:]
```

```python
fig, ax = plt.subplots(2,1,figsize=(15,20))
_ = sb.barplot(data = divf, y="provincia", x= "totalDivorcios", ax=ax[0], hue="provincia")
_ = sb.barplot(data = matf, y="provincia", x= "totalMatrimonios", ax=ax[1], hue="provincia")
_=ax[0].set_ylabel("Provincias")
_=ax[0].set_xlabel("Cantidad de divorciados")
_=ax[0].set_title("Cantidad de divorciados por provincia", {"fontsize":14})
_=ax[1].set_ylabel("Provincias")
_=ax[1].set_xlabel("Cantidad de casados")
_=ax[1].set_title("Cantidad de casados por provincia", {"fontsize":14})
```

Como se puede observar, las provincias "Guayas" y "Pinchincha" son en las que hubo mayor cantidad de personas que contrajeron matrimonio, asi mismo son las provincias en las que mayor cantidad de personas se divorcia en el año.

```python
cantxprov.set_index('provincia', inplace=True)
mf.set_index('provincia', inplace=True)
```

```python
gmatrimonios = gdp.join(mf, how='left')
gdivorcios = gdp.join(cantxprov, how='left')
```

```python
gmatrimonios
```

```python
gdivorcios.reset_index(inplace=True)
gmatrimonios.reset_index(inplace=True)
```

```python
gdivorcios.provincia = gdivorcios.provincia.str.title()
gmatrimonios.provincia = gmatrimonios.provincia.str.title()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(20, 25))
gmatrimonios.plot(column='totalMatrimonios', ax=ax[0], legend=True, cmap="BuPu",
               legend_kwds={'label': "Matrimonios por provincia",
                            'orientation': "horizontal"})
gdivorcios.plot(column='totalDivorcios', ax=ax[1], legend=True, cmap="BuPu",
               legend_kwds={'label': "Divorcios por provincia",
                            'orientation': "horizontal"})

for idx, row in gmatrimonios.iterrows():
    ax[0].annotate(text=row['provincia'], xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                   horizontalalignment='center', fontsize=7, color='grey')

ax[0].set_title("Mapa Coroplético de Matrimonios por Provincia")
ax[0].axis('off')  


for idx, row in gdivorcios.iterrows():
    ax[1].annotate(text=row['provincia'], xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                   horizontalalignment='center', fontsize=7, color='grey')

ax[1].set_title("Mapa Coroplético de Divorcios por Provincia")
ax[1].axis('off')  


# Guardar el resultado en un nuevo archivo GeoJSON
output_path = '../output/totalDivorcios.geojson'
gdivorcios.to_file(output_path, driver='GeoJSON')
```
